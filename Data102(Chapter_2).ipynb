````python
labels = ['Oct','Nov','Dec','Jan','Feb','Mar','Apr','Mei','Jun']
plt.plot(range(len(table_from_drive)),table_from_drive['Amount_Government'],'o--b')
plt.plot(range(len(table_from_drive)),table_from_drive['Amount_Sickleave'],'o--r')  #plot more data to compare
plt.title('Monthly Amount Government')
plt.xlabel('Month')
plt.ylabel('Amount Government')
plt.xticks(range(len(table_from_drive)),labels);
````

````python
labels = ['Oct','Nov','Dec','Jan','Feb','Mar','Apr','Mei','Jun']
plt.plot(range(len(table_from_drive)),table_from_drive['Amount_Government'],'o--b',label='goverment')
plt.plot(range(len(table_from_drive)),table_from_drive['Amount_Sickleave'],'o--r',label='sickleave')  #plot more data to compare
plt.title('Monthly Amount Government vs Amount Sick Leave')
plt.xlabel('Month')
plt.ylabel('Number of People')
plt.xticks(range(len(table_from_drive)),labels)
plt.legend();
````

## BarChart

### Basic BarChart

````python
plt.bar(range(len(table_from_drive)),table_from_drive['Amount_Sickleave'])
plt.title('Monthly Amount Sick Leave by Month')
plt.xlabel('Month')
plt.ylabel('Number of sick leave')
plt.xticks(range(len(table_from_drive)),labels);
````

### Stacked BarChart

````python
plt.bar(range(len(table_from_drive)),table_from_drive['Amount_Sickleave'],color='b',label='sick leave')
plt.bar(range(len(table_from_drive)),table_from_drive['Amount_ Personalleave'],bottom = table_from_drive['Amount_Sickleave'], color='r',label='personal leave')
plt.bar(range(len(table_from_drive)),table_from_drive['Amount_ Vacationleave'],bottom = table_from_drive['Amount_Sickleave']+table_from_drive['Amount_ Personalleave'], color='g',label='vacation leave')
plt.xticks(range(len(table_from_drive)),labels)
plt.grid(True,axis='y')
plt.legend();
````
