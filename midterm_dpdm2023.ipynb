Select the top party from each constituency

```python
top_party_0= result_party[result_party['cons_id']==list(set(result_party['cons_id']))[0]].sort_values(by = 'party_list_vote',ascending = False)[:1]

for cons in list(set(result_party['cons_id']))[1:]:
    top_party_1 = result_party[result_party['cons_id']==cons].sort_values(by = 'party_list_vote',ascending = False)[:1]
    top_party_0 = pd.concat([top_party_0,top_party_1])
```

```python
top_party_0
```

Select the runner up party from each constituency

```python
result_party[result_party['cons_id']==list(set(result_party['cons_id']))[0]].sort_values(by = 'party_list_vote',ascending = False)[1:2]
```

```python
runner_up_0= result_party[result_party['cons_id']==list(set(result_party['cons_id']))[0]].sort_values(by = 'party_list_vote',ascending = False)[1:2]

for cons in list(set(result_party['cons_id']))[1:]:
    runner_up_1 = result_party[result_party['cons_id']==cons].sort_values(by = 'party_list_vote',ascending = False)[1:2]
    runner_up_0 = pd.concat([runner_up_0,runner_up_1])
```
